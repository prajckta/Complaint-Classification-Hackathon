```python
# Create a new DataFrame to store rows with duplicate entries
duplicates_df = pd.DataFrame(columns=df.columns)

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Tokenize the text in the 'narratives' column
    tokens = word_tokenize(row['narratives'].lower())
    
    # Find duplicates within the tokens
    duplicates = set([token for token in tokens if tokens.count(token) > 1])
    
    # If there are duplicates, add the row to the duplicates DataFrame
    if duplicates:
        duplicates_df = duplicates_df.append(row)
        print(f"Duplicate(s) in row {index}: {duplicates}, Entry: {row['narratives']}")

# Save the duplicates DataFrame to a CSV file
duplicates_df.to_csv('duplicates.csv', index=False)

# Remove duplicates from the original DataFrame
df.drop_duplicates(subset=['narratives'], inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame with columns 'product' and 'sentiment'
data = {
    'product': ['credit_card', 'retail_banking', 'credit_reporting', 'mortgages_and_loans', 'debt_collection'],
    'sentiment': ['neutral', 'negative', 'neutral', 'negative', 'negative']
}
df = pd.DataFrame(data)

# Group by product and sentiment, and count occurrences
sentiment_counts = df.groupby(['product', 'sentiment']).size().unstack(fill_value=0)

# Plot the counts
sentiment_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Product Category')
plt.ylabel('Count of Sentiments')
plt.title('Count of Sentiments per Product Category')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sentiment')
plt.show()
```
